```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import SplitClassifier2 as sc
import classification_plotter_2d as cp
import sklearn.datasets as skds
```

## sklearn.datasets.make_blobs

```python
sd = np.random.choice(range(200)) #comment out to keep seed
print(sd) #comment out to keep seed
np.random.seed(sd) #use this to find a seed if you want the same results


X, y = skds.make_blobs(n_features=2, centers=2, n_samples=500)

plt.close() #do before creating a plot to clear it out
plt.figure(figsize=(8, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], marker='o', c='b',
            s=25, edgecolor='k', label=0)
plt.scatter(X[y==1, 0], X[y==1, 1], marker='o', c='r',
            s=25, edgecolor='k', label=1)
plt.legend()
plt.show()
```

```python
split = sc.SplitClassifier(X, y)
split.summary()
```

```python
cp.plot_regions(split, X, y)
```

## sklearn.datasets.make_classification

```python
X, y = skds.make_classification(n_samples = 1000,
                           n_classes = 2,
                           n_features=2, 
                           n_redundant=0, 
                           n_informative=2, 
                           n_clusters_per_class=2,) 

plt.close()
plt.figure(figsize=(8, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], marker='o', c='b',
            s=25, edgecolor='k', label=0)
plt.scatter(X[y==1, 0], X[y==1, 1], marker='o', c='r',
            s=25, edgecolor='k', label=1)
plt.legend()
plt.show()
```

```python
split = sc.SplitClassifier(X, y)
split.summary()
```

```python
cp.plot_regions(split, X, y)
```
