## Creating Histograms

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(loc=8, scale=2, size=100) #8 is mean, 2 is sd, 100 is sample size

plt.hist(x)
plt.show()
```

We can customize our plots by setting additional visual parameters, and by adding labels. 

```python
x = np.random.normal(loc=8, scale=2, size=1000)

plt.hist(x, bins = 20, color='salmon', edgecolor='k', normed=True) #bins is number of bars
plt.xlabel('Observed Value') #adds label to x axis
plt.ylabel('Proportion')#adds label to y axis
plt.title('Sample Drawn from N(8,2)') #adds title to graph
plt.show()
```

There are several [named colors](https://matplotlib.org/gallery/color/named_colors.html) that can be used in Matplotlib plots.

```python
x1 = np.random.normal(loc=6, scale=2, size=1000)
x2 = np.random.normal(loc=12, scale=4, size=1000)
x3 = np.random.normal(loc=20, scale=1, size=1000)

plt.hist(x1, bins = 20, color='salmon', alpha=0.6, normed=True) #alpha is the transparency
plt.hist(x2, bins = 20, color='cornflowerblue', alpha=0.6, normed=True)
plt.hist(x3, bins = 20, color='limegreen', alpha=0.6, normed=True)
plt.xlabel('Observed Value')
plt.ylabel('Proportion')
plt.title('Comparison of Normal Distributions')
plt.show()
```

## Creating Bar Charts

```python
grades = np.random.choice(['A','B','C','D','F'],size=37,p=[0.25,0.3,0.2,0.15,0.1]) #p distributes random by percentages
grade_dist = [sum(grades == 'A'), sum(grades == 'B'), sum(grades == 'C'), sum(grades == 'D'), sum(grades == 'F')]

grade_labels = ['A','B','C','D','F']

plt.bar(range(5), grade_dist) #range tells it how many bars, second parameter says how high each bar is
plt.xticks(range(5), grade_labels)#pass the range(how many labels) and list of labels names
plt.xlabel('Grades')
plt.ylabel('Count')
plt.title('Grade Distribution')
plt.show()
```

## Creating Scatterplots

```python
x = np.random.uniform(0,10,100)
y = 7 + 2 * x + np.random.normal(0,3,100) #normal(mean,std,sampleSize) gives noise to graph; must be same size as x array

plt.scatter(x,y)
plt.show()
```

```python
c1 = np.random.uniform(0,10,2) #find a random point 
c2 = np.random.uniform(0,10,2) #find a random point
sd = np.random.uniform(1,3) #find standard deviation

N = 20 #size of sample

x1 = np.random.normal(c1[0], sd, N) #create array of Xs based on first point
y1 = np.random.normal(c1[1], sd, N) #create array of Ys based on first point
x2 = np.random.normal(c2[0], sd, N)
y2 = np.random.normal(c2[1], sd, N)

plt.scatter(x1, y1) #create scatter plot based on array of X's and Y's
plt.scatter(x2, y2)
plt.show()
```

## Creating Line Plots

```python
ep = np.random.normal(0,1,1000)
x = np.cumsum(ep) #creates array that sums terms of ep. x[1]=ep[1], x[2]=ep[1+2]

plt.plot(x)
plt.show()
```

```python
x = np.linspace(0,5,50) #linspace creates array from 0 to 5 with 50 evenly spaced points
y = np.random.normal(0,2,50)

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,5,50)
y = x**3 - 7 * x**2 + 14 * x

plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,2*3.14159,50)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```
