```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import MLUtilities as mlu
import sklearn.datasets as skds
```

## sklearn.datasets.make_blobs

```python
np.random.seed(34) #so i have the same as instructor
X, y = skds.make_blobs(n_features=2, centers=2, n_samples=500)

plt.close()
plt.figure(figsize=(8, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], marker='o', c='b',
            s=25, edgecolor='k', label=0)
plt.scatter(X[y==1, 0], X[y==1, 1], marker='o', c='r',
            s=25, edgecolor='k', label=1)
plt.legend()
plt.show()
```

```python
split = mlu.SplitClassifier(X, y)
split.summary()
```

```python
mlu.plot_regions(split, X, y)
```

## sklearn.datasets.make_classification

```python
X, y = skds.make_classification(n_samples = 1000,#num of observation
                           n_classes = 2,#num classes to generate
                           n_features=2, #num of features
                           n_redundant=0, #extra features that aren't necessary but included any way
                           n_informative=2, #how many features we want to be valuable
                           n_clusters_per_class=2,) 

plt.close()
plt.figure(figsize=(8, 6))
plt.scatter(X[y==0, 0], X[y==0, 1], marker='o', c='b',
            s=25, edgecolor='k', label=0)
plt.scatter(X[y==1, 0], X[y==1, 1], marker='o', c='r',
            s=25, edgecolor='k', label=1)
plt.legend()
plt.show()
```

```python
split = mlu.SplitClassifier(X, y)
split.summary()
```

```python
mlu.plot_regions(split, X, y)
```
